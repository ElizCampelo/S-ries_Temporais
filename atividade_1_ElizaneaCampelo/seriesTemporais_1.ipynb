```python
# Carregar os dados
import pandas as pd
import statistics
import matplotlib.pyplot as plt
beer_production_data = pd.read_csv('C:\\Users\\eliz1\Desktop\\cesarSschool2024\\SeriesTemporais\\atividade_1_ElizaneaCampelo\\bearq.txt', delimiter='\t')
river_flows_data = pd.read_csv('C:\\Users\\eliz1\\Desktop\\cesarSschool2024\\SeriesTemporais\\atividade_1_ElizaneaCampelo\\coloradoRiver.txt', delimiter='\t')
temperature_data = pd.read_csv('C:\\Users\\eliz1\\Desktop\\cesarSschool2024\\SeriesTemporais\\atividade_1_ElizaneaCampelo\\temperature.txt', delimiter='\t')
```

```python
# Análise estatística descritiva p
print("Estatísticas Descritivas para Quarterly beer production in Australia:")
print(beer_production_data.describe())
print("Estatísticas Descritivas Colorado River Lees Ferry :")
print(river_flows_data.describe())
print("Estatísticas Descritivas para temperature:")
print(temperature_data.describe())
```

```python
# Plotar os dados da produção trimestral de cerveja
plt.plot(range(1, len(beer_production_data) + 1), beer_production_data, marker='o', linestyle='-')
plt.title("Quarterly beer production in Australia")
plt.xlabel("Quarter")
plt.ylabel("Production (megalitres)")
plt.grid(True)
plt.show()

#plotar os dados da série temporal "Monthly River Flows (Colorado River Lees Ferry)
plt.plot(range(1, len(river_flows_data) + 1), river_flows_data, marker='o', linestyle='-')
plt.title("Monthly River Flows (Colorado River Lees Ferry)")
plt.xlabel("Month")
plt.ylabel("River Flows")
plt.grid(True)
plt.show()

# Plotar os dados da temperatura ao longo do tempo
plt.plot(range(1, len(temperature_data) + 1), temperature_data, marker='o', linestyle='-')
plt.title("Monthly air temperature (Nottingham, 1920–1939)")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.grid(True)
plt.show()
```
