```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

```

```python
# Carregar dados
file_path = 'C:\\Users\eliz1\\Desktop\\cesarSschool2024\\SeriesTemporais\\atividade_1_ElizaneaCampelo\\bearq.txt'
data = pd.read_csv(file_path, header=None, delimiter='\t', names=['actual'])

# Visualizar os primeiros dados
print(data.head())
```

```python
# Estatísticas descritivas
print(data.describe())
```

```python
# Plotar a série temporal
plt.figure(figsize=(12, 6))
plt.plot(data['actual'], label='Actual')
plt.title('Time Series Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
# Plotar ACF e PACF
fig, ax = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(data['actual'], lags=40, ax=ax[0])
plot_pacf(data['actual'], lags=40, ax=ax[1])
plt.show()
```

```python
# Teste de Dickey-Fuller Aumentado
adf_result = adfuller(data['actual'])

# Resultados do teste ADF
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
print('Critical Values:', adf_result[4])
```
